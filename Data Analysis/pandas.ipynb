## Data - Manipulation using pandas

```python
df = pd.read_csv('housing_data.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.fillna(0)
```

```python
# filling missing values with mean of the column
df['RM_fillNA'] = df['RM'].fillna(df['RM'].mean())
```

```python
df
```

```python
df.dtypes
```

```python
# renaming the column
df = df.rename(columns={'RM':'RM_Value'})
df.head()
```

```python
# changing datatype
df['RM_new'] = df['RM_Value'].fillna(df['RM_Value'].mean()).astype(int)
df.head()
```

```python
df['New CRIM'] = df['CRIM'].apply(lambda x:x*2)
df.head()
```

```python
# Data Aggregating and Grouping
df.head()
```

```python
grouped_mean = df.groupby('RAD')['AGE'].mean()
print(grouped_mean)
```

```python
grouped_mean = df.groupby('CHAS')['AGE'].mean()
print(grouped_mean)
```

```python
grouped_sum = df.groupby(['CHAS','RAD'])['MEDV'].sum()
print(grouped_sum)
```

```python
# aggregrate multiple functions
grouped_agg = df.groupby(['CHAS','RAD'])['AGE'].agg(['mean','sum','count'])
grouped_agg
```

```python
# Merging and joining DataFrames
df1 = pd.DataFrame({'Key':['A','B','C'],'Value1':[1,2,3]})
df2 = pd.DataFrame({'Key':['A','B','C','D'],'Value2':[4,5,6,7]})
```

```python
df1
```

```python
df2
```

```python
# Merge DataFrame on the 'key columns'
pd.merge(df1,df2,on="Key",how='inner')
```

```python
pd.merge(df1,df2,on="Key",how='outer')
```

```python
pd.merge(df1,df2,on="Key",how='left')
```

```python
pd.merge(df1,df2,on="Key",how='right')
```
