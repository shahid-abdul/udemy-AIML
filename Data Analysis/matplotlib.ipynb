```python
import matplotlib.pyplot as plt
```

## line plot

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Basic Plot')
plt.show()
```

## customised line plot

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

plt.plot(x,y,color='red',linestyle='--',marker='o',linewidth=3,markersize=9)
plt.grid(True)
```

## Multiple Plots

```python
# sample data

x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [2,4,6,8,10]

plt.figure(figsize=(9,5))

plt.subplot(2,2,1)
plt.plot(x,y1,color='green')
plt.title('Plot 1')
plt.grid(True)

plt.subplot(2,2,2)
plt.plot(y1,x,color='red')
plt.title('Plot 2')
plt.grid(True)

plt.subplot(2,2,3)
plt.plot(x,y2,color='blue')
plt.title('Plot 3')
plt.grid(True)

plt.subplot(2,2,4)
plt.plot(y2,x,color='orange')
plt.title('Plot 4')
plt.grid(True)
```

## Bar plot

```python
categories = ['A','B','C','D','E','F']
values = [2,5,8,3,9,4]

plt.bar(categories,values,color='purple')
plt.xlabel('Categories')
plt.ylabel('values')
plt.title('Bar Plot')
```

## Histograms

```python
data = [1,2,3,2,5,1,4,9,5,2,1,7,4,3,7,8,3,9,2,8,3,6,4,6,3,7]

plt.hist(data,bins=5,color='orange',edgecolor='black')
```

## scatter plot

```python
x = [1,2,3,4,5]
y = [2,3,4,5,6]

plt.scatter(x,y,color='green',marker='o')
```

## pie chart

```python
labels = ['A','B','C','D']
sizes = [30,20,40,10]
colors = ['gold','yellowgreen','lightcoral','lightskyblue']
explode = (0.2,0,0,0)

plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct="%1.1f%%",shadow=True)
```

```python
import pandas as pd
df = pd.read_csv("car data.csv")
df.head()
```

```python
df.info()
```

```python
total_km_driven = df.groupby('Fuel_Type')['Kms_Driven'].sum()
print(total_km_driven)
```

```python
total_km_driven.plot(kind='bar',color='teal')
```

```python
total_km_driven = df.groupby(['Fuel_Type','Year'])['Kms_Driven'].sum()
print(total_km_driven)
```

```python
total_km_driven.plot(kind='bar',color='teal')
```

```python
total_km_driven = df.groupby(['Year'])['Kms_Driven'].sum()
print(total_km_driven)
```

```python
total_km_driven.plot(kind='bar',color='teal')
```

```python
profit = df.groupby('Transmission').apply(lambda x: (x['Present_Price'] - x['Selling_Price']).sum())
profit.plot(kind='bar',color='teal')
```

```python
profit = df.groupby('Seller_Type').apply(lambda x: (x['Present_Price'] - x['Selling_Price']).sum()).reset_index()
profit.plot(kind='bar',color='teal')
```

```python
profit = df.groupby('Fuel_Type').apply(lambda x: (x['Present_Price'] - x['Selling_Price']).sum())
profit.plot(kind='pie',explode=(0.1,0.1,0.1))
```
